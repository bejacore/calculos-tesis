```python
from astropy.table import Table
from astropy.table import unique
import pandas as pd
```

## Creación de tablas

```python
clusters = Table.read('../data/raw/clusters.fit')
```

```python
members = Table.read('../data/raw/members.fit')
```

## Selecciona solo los datos de los cúmulos abiertos y globulares

```python
mask = (clusters['Type'] == 'o') | (clusters['Type'] == 'g')
clusters = clusters[mask]
```

## Limpiar miembros repetidos

```python
clean_members = unique(members, keys='GaiaDR3')
```

## Pasar de Astropy Table a Pandas

```python
clusters.to_pandas().to_csv('../data/processed/clusters.csv', index=False)
clean_members.to_pandas().to_csv('../data/processed/members.csv', index=False)
```

```python
clusters = pd.read_csv('../data/processed/clusters.csv')
members = pd.read_csv('../data/processed/members.csv')
```

## Formatear strings

Los strings en las tablas de Astropy eran bytes objects, al convertir esas tablas en formato csv los strings tenían los símbolos `b'` y `'` al inicio y final de estos.

```python
clusters['Name'] = clusters['Name'].str.strip("b'")
clusters['Type'] = clusters['Type'].str.strip("b'")
```

Se filtran solo los miembros de los cúmulos abiertos y globulares, dejado de lado asterismos y otros objetos.

```python
members['Name'] = members['Name'].str.strip("b'")
```

```python
valid_clusters = clusters['Name'].unique()
members = members[members['Name'].isin(valid_clusters)]
```

```python
members = members.reset_index(drop=True)
```

```python
clusters.to_csv('../data/processed/clusters.csv', index=False)
members.to_csv('../data/processed/members.csv', index=False)
```
