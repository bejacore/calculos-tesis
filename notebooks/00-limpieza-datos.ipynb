```python
from astropy.table import Table
from astropy.table import unique
import pandas as pd
```

## Creación de tablas

```python
clusters = Table.read('../data/raw/clusters.fit')
```

```python
members = Table.read('../data/raw/members.fit')
```

## Selección de open clusters

Solo se seleccionan estos tipos de cúmulos, ya que son estos en los que se centran el catálogo de Hunt & Reffert.

```python
mask = clusters['Type'] == 'o'
open_clusters = clusters[mask]
```

## Limpiar miembros repetidos

```python
clean_members = unique(members, keys='GaiaDR3')
```

## Pasar de Astropy Table a Pandas

```python
open_clusters.to_pandas().to_csv('../data/processed/open_clusters.csv', index=False)
clean_members.to_pandas().to_csv('../data/processed/members.csv', index=False)
```

```python
open_clusters = pd.read_csv('../data/processed/open_clusters.csv')
members = pd.read_csv('../data/processed/members.csv')
```

## Formatear strings

Los strings en las tablas de Astropy eran bytes objects, al convertir esas tablas en formato csv los strings tenían los símbolos `b'` y `'` al inicio y final de estos.

```python
open_clusters['Name'] = open_clusters['Name'].str.strip("b'")
open_clusters['Type'] = open_clusters['Type'].str.strip("b'")
```

```python
members['Name'] = members['Name'].str.strip("b'")
```

```python
open_clusters.to_csv('../data/processed/open_clusters.csv', index=False)
members.to_csv('../data/processed/members.csv', index=False)
```
